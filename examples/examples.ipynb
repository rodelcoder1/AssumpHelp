```python
import numpy as np
from sklearn.linear_model import LinearRegression
import assumphelper as ah

# Reproducibility
np.random.seed(42)

# Generate linear data
n = 100
X = np.linspace(1, 100, n).reshape(-1, 1)
y = 3 * X.ravel() + 10 + np.random.normal(0, 5, n)

# Model
model = LinearRegression()

# Linearity
linearity = ah.Linearity(model, X, y)
linearity.fit()
linearity.test()
linearity.plot()
```

```python
# Normality
normality = ah.Normality(model, X, y)
normality.fit()
normality.test()
normality.plot()
```

```python
# Homoscedasticity
homoscedasticity = ah.Homoscedasticity(model, X, y)
homoscedasticity.fit()
homoscedasticity.test()
homoscedasticity.plot()
```

```python
# Independence
independence = ah.Independence(model, X, y)
independence.fit()
independence.plot()
```
